# Otros operadores en Python...

**Operadores de Comparación**

Son símbolos que nos permiten comparar dos valores y obtener como resultado True o False

| Operador | Significado | Ejemplo | Resultado |
|----------|-------------|---------|-----------|
| `==` | Igual a | `5 == 5` | `True` |
| `!=` | Diferente de | `5 != 3` | `True` |
| `>` | Mayor que | `10 > 5` | `True` |
| `<` | Menor que | `3 < 2` | `False` |
| `>=` | Mayor o igual | `18 >= 18` | `True` |
| `<=` | Menor o igual | `15 <= 10` | `False` |

**Importante:**  
- `=` sirve para **asignar** valores (`edad = 18`)  
- `==` sirve para **comparar** valores (`edad == 18`)  
- Al comparar textos, se distinguen mayúsculas: `"Ana" == "ana"` → `False`

*Recuerda que...* Debes usar `.lower()` para comparar textos sin importar mayúsculas:  
`nombre.lower() == "ana"`

```python
# OPERADORES DE COMPARACIÓN

print("Comparemos numeros...")
edad = 18

print("edad es igual a 18?", edad == 18)
print("edad es diferente de 21?", edad != 21)
print("edad es mayor que 15?", edad > 15)
print("edad es menor que 25?", edad < 25)
print("edad es mayor o igual a 18?", edad >= 18)
print("edad es menor o igual a 17?", edad <= 17)

print("")


# Lo mismo pero pidiendo dos valores:

num1 = float(input("Ingresa el primer numero: "))
num2 = float(input("Ingresa el segundo numero: "))

print("")
print(num1, "==", num2, "->", num1 == num2)
print(num1, "!=", num2, "->", num1 != num2)
print(num1, ">", num2, "->", num1 > num2)
print(num1, "<", num2, "->", num1 < num2)
print(num1, ">=", num2, "->", num1 >= num2)
print(num1, "<=", num2, "->", num1 <= num2)
```

**Operadores Lógicos**

Permiten combinar dos o más condiciones para tomar decisiones más complejas.


| Operador | Nombre | ¿Cuándo es True? | Ejemplo |
|----------|--------|------------------|---------|
| `and` | Y lógico | Cuando **ambas** condiciones son verdaderas | `edad >= 18 and tiene_licencia` |
| `or` | O lógico | Cuando **al menos una** condición es verdadera | `es_estudiante or es_jubilado` |
| `not` | Negación | Invierte el valor: True → False, False → True | `not esta_vacio` |

Por ejemplo:

*"Puedo salir **SI** tengo permiso  **Y** terminé mi tarea"*  
Si falta una de las dos, no puedo salir

```python
# OPERADORES LÓGICOS: and, or, not
# and
# Quiero saber si puedo conducir o no

tiene_18_años = True
tiene_licencia = False

if tiene_18_años and tiene_licencia:
    print("Puede conducir")
else:
    print("No puede conducir")

```

```python
# or
#Quiero saber si se me aplica descuento
es_estudiante = False
es_adulto_mayor = True

if es_estudiante or es_adulto_mayor:
    print("Aplica para descuento")
else:
    print("No hay descuento")

```

```python
#Not
#Supongamos que tengo una lista de compras
esta_vacia = False

if not esta_vacia:
    print("Hay elementos en mi lista")
else:
    print("La lista está vacía")

```

```python
# Acceso a una red social
#Debes ser mayor de edad y aceptar los terminos y condiciones

edad = int(input("¿Cuántos años tienes? "))
terminos = input("¿Aceptas los términos? (si/no): ").lower()

if edad >= 18 and terminos == "si":
    print("Bienvenido")
else:
    if edad < 18:
        print("Debes tener al menos 18 años")
    if terminos != "si":
        print("Debes aceptar los términos para continuar")
```
