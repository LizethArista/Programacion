# Listas de Objetos

En Python, una lista puede almacenar cualquier tipo de dato, incluidas **instancias de clases (objetos)**. Esto es muy útil para manejar información más aplicadas: estudiantes, productos, figuras geométricas, empleados, etc.


```python
# Definimos la calse Estudiante
class Estudiante:
    def __init__(self, nombre, edad, promedio):
        self.nombre = nombre
        self.edad = edad
        self.promedio = promedio

    def __str__(self):
        return f"{self.nombre} ({self.edad} años),  Promedio: {self.promedio}"

    def obtener_promedio(self):
        return self.promedio
```

## ¿Qué acabamos de definir?
- `__init__`: Constructor que inicializa los atributos de cada objeto.
- `__str__`: Método especial que define cómo se verá el objeto cuando lo imprimamos.
- Cada vez que llamamos `Estudiante(...)` creamos un **objeto independiente** en memoria.

```python
# Creamos varios objetos Estudiante
e1 = Estudiante("Ana", 20, 9.5)
e2 = Estudiante("Luis", 22, 8.0)
e3 = Estudiante("María", 19, 9.8)

# Los guardamos en una lista
grupo = [e1, e2, e3]

print("Lista de objetos:")
for estudiante in grupo:
    print(estudiante)
```

## Operaciones comunes con listas de objetos
Las listas de objetos heredan todos los métodos de las listas estándar de Python: `append()`, `len()`, slicing (extracción de [inicio:parada:paso]),  `sorted()` (para ordenar), etc. También puedes modificar los atributos de los objetos directamente a través de la lista.



```python
ordenado = sorted(grupo, key=Estudiante.obtener_promedio, reverse=True)
for e in ordenado:
    print(e)
```

sorted() solo pide que key sea cualquier función que reciba 1 elemento y devuelva el valor de ordenamiento. lambda es solo una forma compacta de crear esa función.

key = lambda [variable]: [valor_a_comparar]

lambda indica que es una función anónima (de una sola línea).

[variable] es el nombre que tú quieras (x, e, alumno, item...)

[valor_a_comparar] es lo que sorted() usará para ordenar. Ahí puedes poner lo que necesites.

```python
# Agrego un nuevo estudiante
grupo.append(Estudiante("Carlos", 21, 7.5))
print(f" Total de estudiantes: {len(grupo)}\n")

# Filtro a los estudiantes con promedio >= 9 (comprensión de listas)
destacados = [e for e in grupo if e.promedio >= 9.0]
print("Estudiantes destacados:")
for d in destacados:
    print(d)

# Ordeno por promedio (de mayor a menor)
grupo_ordenado = sorted(grupo, key=lambda x: x.promedio, reverse=True)
print("\n Ordenados por promedio:")
for e in grupo_ordenado:
    print(e)
```

```python
# Modifico un objeto directamente desde la lista
print(f"Antes: {grupo[0]}")
grupo[0].promedio = 10.0
print(f"Después: {grupo[0]}")

# Verifico que el cambio se refleja en la lista original
print("\n Lista actualizada:")
for e in grupo:
    print(e)
```

## Reasignar vs Mutar
Si haces `e = nuevo_objeto` dentro de un `for`, **solo cambias la variable local**, no el objeto que vive dentro de la lista. Para modificar el objeto original, debes actuar sobre sus atributos o usar índices.

```python
# Solo reasigna la variable local `e`, la lista NO cambia
for e in grupo:
    e = Estudiante(e.nombre, e.edad, e.promedio + 0.5)

print("Lista incorrecta (no cambia):")
print([str(e) for e in grupo])

# Modificamos el atributo del objeto referenciado
for e in grupo:
    e.promedio += 0.5  # Sí muta el objeto original

print("\n Lista correcta:")
print([str(e) for e in grupo])
```
