#EJEMPLO DE CADENAS

Una cadena es una secuencia de carácteres utilizada para representar texto

Python nos permite crear cadenas de diferentes maneras, dependiendo lo que necesitemos

¿Cómo podemos formar cadenas?
1. Con comillas simples y dobles
2. Cadenas multilínea
3. Cadenas "crudas" (Rax string)
4. f- string (cadenas formateadas)


```python
# Comillas simples y dobles
# Podemos usar cualquiera, para definir texto simple

texto_simple = 'Hola Mundo'
print(texto_simple)
texto_doble = "Hola Mundo"
print(texto_doble)

#Usa comillas simples si tu código contiene comillas dobles
texto2 = 'Él dijo: "Me gusta programar"'
print(texto2)
```

```python
# Cadenas multilínea

# Si usamos tres comillas simples o tres dobles, podemos escribir texto
# en varias líneas (para colocar bloques de texto largos)

poema = '''
Rosa es roja,
violeta es azul,
el azúcar es dulce,
y tú también.

'''
print(poema)

#Otro ejemplo... dirección
direccion = """
Calle 123,
Departamento 48,
Ciudad de México,
México
"""
print(direccion)
```

```python
#Raw strings

#Se usan poniendo una 'r' antes de las comillas
# Esto le indica a Python que ignore los carácteres especiales como (\n)
# Es útil para rutas de archivos


#Sin la 'r' Python interpreta \n como un salto de línea

ruta_normal = "C:\nuevo\trabajo"
print("Mi ruta Normal:", ruta_normal)

#Con la 'r' Python lo trata como un texto
ruta_cruda = r"C:\nuevo\trabajo"
print("Mi ruta Cruda:", ruta_cruda)
```

```python
# f strings

#Es la forma recomendada de insertar variables dentro de un texto
nombre = "Ana"
edad =25
altura = 1.65

#Primera forma
mensaje1 = "Hola, te llamas " + nombre + " y tienes " + str(edad) + " años"

#Usando f string
mensaje2 = f"Hola te llamas {nombre} y tienes {edad} años"

#con esto también podemos realizar opreaciones dentro del paréntesis
doble_edad= f"El doble de tu edad es {edad*2}"

print(mensaje1)
print(mensaje2)
print(doble_edad)
```

#Posiciones de cadenas

Tenemos:
- Indices positivos (comienzan desde cero desde la izquierda)
- Indices negativos (comienzan en -1 desde la derecha)
- Slicing (podemos extraer partes de la cadenas con [inicio: fin]


Por ejemplo...

Cadena:   P  Y  T  H  O  N



```python
# Accediendo a caracteres por su posición

palabra = "Python"

# Índice positivo (desde el inicio, empieza en 0)
print("Primera letra:", palabra[0]) # P
print("Segunda letra:", palabra[1]) # y
print("Última letra:", palabra[5]) # n

# Índice negativo (desde el final, empieza en -1)
print("Última letra (negativo):", palabra[-1]) # n
print("Primera letra (negativo):", palabra[-6]) # P

# ¿Qué pasa si uso un índice que no existe?
# print(palabra[10]) # Esto daría error: IndexError

```

```python
# SLICING CON PASO (STEP)
# Sintaxis: cadena[inicio:fin:paso]

palabra = "Python"

print("Cada 2 letras:", palabra[0:6:2]) # Pto (salta de 2 en 2)
print("Cada 2 letras (simplificado):", palabra[::2]) # Pto

# Para invertir una cadena
print("Cadena invertida:", palabra[::-1]) # nohtyP

# Saltar letras desde el final
print("De atrás hacia adelante:", palabra[::-2]) # nhy
```

```python
# Ejemplo con una frase completa

frase = "Hola Mundo Python"

print("Frase completa:", frase)
print("Longitud:", len(frase)) # 18 caracteres (incluye espacios)
print("Primera palabra:", frase[0:4]) # Hola
print("Segunda palabra:", frase[5:10]) # Mundo
print("Tercera palabra:", frase[11:]) # Python

# Acceder a espacios
print("Posición del primer espacio:", frase[4]) # ' ' (espacio vacío)
```

```python
# Funciones útiles para trabajar con posiciones

frase = "Hola Mundo Python"

# Longitud de la cadena
print("Longitud:", len(frase))

# Encontrar la posición de una subcadena
print("Posición de 'Mundo':", frase.find("Mundo")) # 5

# Verificar si existe una subcadena
print("¿Existe 'Python'?:", "Python" in frase) # True
print("¿Existe 'Java'?:", "Java" in frase) # False

# Contar cuántas veces aparece una letra
print("Cuántas 'o':", frase.count("o")) # 2
```
