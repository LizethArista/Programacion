# Ejercicio 3: Recorriendo Cadenas con `while`

**Instrucción:**
Crea una función `contar_vocales` que reciba un texto. Usa un `while` para recorrer el string carácter por carácter y cuenta cuántas vocales hay (a, e, i, o, u).

MÁS ADELANTE VEREMOS QUE PARA ESTO PODEMOS USAR `for`.

````python
# EJERCICIO 3: Contador de Vocales

def contar_vocales(texto):
    vocales = "aeiouAEIOU"
    contador = 0
    indice = 0

    while indice < len(texto):
        if texto[indice] in vocales:
            contador += 1
        indice += 1

    return contador
contar_vocales("hOLA")
````

# Otra función que podría ser de utilidad...
La palabra **assert** significa "afirmar" o "asegurar". En programación, se utiliza como una herramienta de depuración (debugging) y pruebas.


Funciona como un guardia de seguridad en tu código:

Verifica una condición: Le das una expresión que debe ser verdadera (True).

- Si es Verdadera: El programa ignora el assert y continúa ejecutándose como si nada.

- Si es Falsa: El programa se detiene inmediatamente y lanza un error llamado AssertionError.


En el ejercicio de contar_vocales, usamos assert para hacer pruebas automáticas. En lugar de ejecutar la función, mirar el resultado, calcular mentalmente si es correcto y luego probar con otro texto, el assert lo hace por nosotros:

1-. "Afirmo que contar_vocales('Hola') debe ser igual a 2".

Si tu función devuelve 2: No pasa nada (silencio = éxito)

Si tu función devuelve 3: ¡ERROR! El programa se detiene y te avisa que algo está mal

````python
# Para hacer pruebas automáticas
assert contar_vocales("Hola") == 2
assert contar_vocales("Python") == 1
assert contar_vocales("Escuela") == 4
````
