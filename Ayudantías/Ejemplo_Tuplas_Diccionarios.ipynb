```python
class GestorEstudiantes:
    """
    Clase que administra información de estudiantes utilizando:
    - Tuplas para datos fijos e inmutables
    - Diccionarios para registros actualizables
    """

    def __init__(self, escala_calificaciones: tuple):
        """
        Inicializacion
        param escala_calificaciones: Tupla con los límites de calificación (min, max).
                                      Se usa tuple porque es una regla fija que no debe cambiar
        """
        # Validamos que la escala tenga exactamente 2 valores
        if len(escala_calificaciones) != 2:
            raise ValueError("La escala debe contener exactamente dos valores: (mínimo, máximo).")

        self.escala = escala_calificaciones  # Tupla inmutable
        self.estudiantes = {}                # Diccionario mutable para almacenar registros

    def agregar_estudiante(self, id_estudiante: str, datos_fijos: tuple, contacto: dict):
        """
        Registra un nuevo estudiante
        param id_estudiante: Clave única (str)
        param datos_fijos: Tupla con (nombre, fecha_nacimiento, carrera)
        param contacto: Diccionario con email, teléfono, etc
        """
        # Evitamos sobrescribir un ID ya existente
        if id_estudiante in self.estudiantes:
            print(f"El ID '{id_estudiante}' ya está registrado.")
            return

        # Desempaquetamos la tupla para validar/verificar datos
        nombre, fecha, carrera = datos_fijos

        # Creamos el registro interno combinando tupla y diccionario
        self.estudiantes[id_estudiante] = {
            "datos_fijos": datos_fijos,   # Se guarda la tupla tal cual
            "contacto": contacto,         # Diccionario mutable
            "notas": {},                  # Diccionario vacío que se llenará después
            "activo": True                # Estado booleano
        }
        print(f" Estudiante '{nombre}' registrado correctamente")

    def actualizar_nota(self, id_estudiante: str, materia: str, nota: float):
        """
        Actualiza o agrega una nota en el diccionario interno del estudiante
        """
        if id_estudiante not in self.estudiantes:
            raise KeyError(f"Estudiante con ID '{id_estudiante}' no encontrado")

        estudiante = self.estudiantes[id_estudiante]

        # Validamos que la nota esté dentro de la escala fija (tupla)
        min_nota, max_nota = self.escala
        if not (min_nota <= nota <= max_nota):
            raise ValueError(f"La nota debe estar entre {min_nota} y {max_nota}.")

        # Los diccionarios permiten agregar/actualizar claves dinámicamente
        estudiante["notas"][materia] = nota
        print(f"Nota de {materia} actualizada a {nota} para {estudiante['datos_fijos'][0]}")

    def calcular_promedio(self, id_estudiante: str) -> float:
        """
        Calcula el promedio de notas. Devuelve 0 si no hay notas registradas
        """
        if id_estudiante not in self.estudiantes:
            raise KeyError("ID no encontrado.")

        notas = self.estudiantes[id_estudiante]["notas"]
        if not notas:
            return 0.0

        # sum() y len() funcionan con diccionarios de valores numéricos
        return sum(notas.values()) / len(notas)

    def obtener_resumen(self, id_estudiante: str) -> tuple:
        """
        Devuelve un resumen inmutable (tupla) con información
        Útil para exportar datos sin riesgo de modificación accidental
        """
        if id_estudiante not in self.estudiantes:
            raise KeyError("ID no encontrado.")

        est = self.estudiantes[id_estudiante]
        nombre = est["datos_fijos"][0]
        promedio = self.calcular_promedio(id_estudiante)
        total_materias = len(est["notas"])

        # Retornamos una tupla
        return (id_estudiante, nombre, promedio, total_materias, est["activo"])

    def cambiar_estado(self, id_estudiante: str, activo: bool):
        """
        Modifica el estado del estudiante. Muestra mutabilidad vs inmutabilidad.
        """
        self.estudiantes[id_estudiante]["activo"] = activo
        print(f"Estado de {id_estudiante} cambiado a {'ACTIVO' if activo else 'INACTIVO'}.")
```

```python
# Instanciamos la clase con una tupla fija de escala de calificaciones (0 a 10)
gestor = GestorEstudiantes(escala_calificaciones=(0, 10))

# Agregamos estudiantes usando tuplas para datos fijos y diccionarios para contacto
gestor.agregar_estudiante(
    id_estudiante="E001",
    datos_fijos=("Ana García", "2000-03-15", "Ingeniería"),  # Tupla
    contacto={"email": "ana@mail.com", "telefono": "555-0101"}  # Diccionario
)

gestor.agregar_estudiante(
    id_estudiante="E002",
    datos_fijos=("Luis Pérez", "1999-11-22", "Medicina"),
    contacto={"email": "luis@mail.com", "telefono": "555-0202"}
)

print(gestor.obtener_resumen("E001"))

print(gestor.obtener_resumen("E002"))
```

```python

# Actualizamos notas
gestor.actualizar_nota("E001", "Matemáticas", 8.5)
gestor.actualizar_nota("E001", "Programación", 9.2)
gestor.actualizar_nota("E002", "Biología", 7.0)


print(gestor.obtener_resumen("E001"))

print(gestor.obtener_resumen("E002"))
```

```python
# Calculamos promedios
prom_ana = gestor.calcular_promedio("E001")
print(f"\n Promedio de Ana: {prom_ana:.2f}")

print(gestor.obtener_resumen("E001"))

print(gestor.obtener_resumen("E002"))

```

```python
# Obtenemos resumen como tupla (datos seguros e inmutables)
resumen = gestor.obtener_resumen("E001")
print(f"\n Resumen (tupla): ID={resumen[0]}, Nombre={resumen[1]}, Promedio={resumen[2]:.2f}, Materias={resumen[3]}")

```

```python
print("\nPrueba de inmutabilidad de la tupla 'escala':")
try:
    gestor.escala[0] = 1.0  # Esto lanzará TypeError
except TypeError as e:
    print(f"Error esperado: {e}  Las tuplas no se pueden modificar")

print("\nPrueba de mutabilidad del diccionario 'contacto':")
gestor.estudiantes["E001"]["contacto"]["telefono"] = "555-9999"
print(f"Teléfono actualizado: {gestor.estudiantes['E001']['contacto']['telefono']}... Los diccionarios sí permiten cambios")

# para dar de baja a un estudiante
gestor.cambiar_estado("E002", False)
resumen_luis = gestor.obtener_resumen("E002")
print(f"\n resumen actualizado de Luis: {resumen_luis}")
```

### Diccionario de Notas

1-.  Comienza vacío `{}`

2-.  Se crean claves nuevas automáticamente al agregar materias

3-. Si una materia ya existe, **se sobrescribe** el valor (mutabilidad)

4-. No requiere definir estructura previa ni tamaño fijo


```python
import json

gestor = GestorEstudiantes(escala_calificaciones=(0.0, 10.0))
gestor.agregar_estudiante(
    id_estudiante="E001",
    datos_fijos=("Ana García", "2000-03-15", "Ingeniería"),
    contacto={"email": "ana@mail.com"}
)

# ESTADO INICIAL del diccionario de notas
print("ESTADO INICIAL:")
print(json.dumps(gestor.estudiantes["E001"]["notas"], indent=4, ensure_ascii=False))

```

```python
#  PRIMERA NOTA
print("\n AGREGANDO: Matemáticas = 8.5")
gestor.actualizar_nota("E001", "Matemáticas", 8.5)
print(" ESTADO DESPUÉS DE LA PRIMERA MATERIA:")
print(json.dumps(gestor.estudiantes["E001"]["notas"], indent=4, ensure_ascii=False))

```

```python
# AGREGAR SEGUNDA NOTA (nueva clave)
print("\n AGREGANDO: Programación = 9.2")
gestor.actualizar_nota("E001", "Programación", 9.2)
print(" ESTADO CON DOS MATERIAS:")
print(json.dumps(gestor.estudiantes["E001"]["notas"], indent=4, ensure_ascii=False))

```

```python
# ACTUALIZAR NOTA EXISTENTE (sobrescritura por clave)
print("\n ACTUALIZANDO: Matemáticas = 9.5 (sobrescribe el 8.5 anterior)")
gestor.actualizar_nota("E001", "Matemáticas", 9.5)
print(" ESTADO FINAL:")
print(json.dumps(gestor.estudiantes["E001"]["notas"], indent=4, ensure_ascii=False))

```

```python
# Estructura del estudiante
print("\n ESTRUCTURA COMPLETA DEL ESTUDIANTE:")
print(json.dumps(gestor.estudiantes["E001"], indent=4, ensure_ascii=False))
```
