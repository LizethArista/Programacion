# INTRODUCCIÓN A PYTHON

**Buenas prácticas de programación**


```python
#=============================================
# Vamos a crear una pequeña función inicial...
# ============================================

def main():
  """
  Esta una función que me va a pedir datos sobre las personas
  para poder tener un registro

  """
  #ENTRADAS (inputs)

  nombre_usuario = input("escribe tu nombre:")
  edad = input("escribe tu edad en años:")
  altura = input ("esribe tu altura: ")
  programador = input("Te gusta programar? (si/no): ")

  # TIPOS DE DATOS
  edad = int(edad)
  altura = float(altura)
  programador = programador.lower() == "si"
                                        # .lower() pasa el texto a minúsculas
                                        # == es para comparar...

  #SALIDAS

  print("ficha de usuario")
  print("="*30)

  print(f"Nombre: {nombre_usuario}")
  print(f"Edad: {edad}")
  print(f"Altura: {altura}")
  print(f"Es programador: {programador}")

  anio_nacimiento = 2026 - edad
  print(f"Año de nacimiento: {anio_nacimiento}")

main ()
```

**Buenas Prácticas de Programación**

- Nombres descriptivos: En lugar de usar x, y o n, usamos nombre_usuario, edad_texto. Esto hace que el código se lea como inglés (o español) simple. Cualquiera que lo lea sabrá qué contiene esa variable.


- Funciones (def main): No escribimos todo el código "suelto". Lo metemos dentro de una función llamada main. Esto ayuda a organizar el pensamiento lógico.


- Comentarios: Las líneas que empiezan con # no se ejecutan. Sirven para explicarle a otros humanos (o a tu yo del futuro) qué hace ese bloque de código.

**Entradas y Salidas**

- input("Mensaje"): Es la puerta de entrada. El programa se detiene y espera a que el usuario escriba algo y presione Enter

NOTA: Python trata todo lo que escribe el usuario como texto (cadenas de caracteres), aunque escriban un número

- print("Mensaje"): Es la puerta de salida. Muestra información en la pantalla.


**Tipos de Datos y Conversión**

- String (str): Texto. Ejemplo: "Juan". Viene directamente del input

- Integer (int): Números enteros. Ejemplo: 25. No puedes hacer matemáticas con un texto, por eso usamos int(edad_texto) para convertir el texto "25" al número 25

- Float (float): Números con decimales. Ejemplo: 1.75. Usamos float() para la altura

- Boolean (bool): Solo tiene dos valores: True (Verdadero) o False (Falso). En el código, convertimos la respuesta "si/no" en un valor lógico verdadero/falso
