> **¿Qué es una clase?**  
> Una clase es un **"plano" o "molde"** para crear objetos.

Define:  
> • **Atributos:** Datos que almacena cada objeto (características)  
> • **Métodos:** Acciones que puede realizar ese objeto

> **Estructura**  
> ```python
> class NombreClase:
>     def __init__(self, param1, param2):
>         self.param1 = param1  # Atributo de instancia
>         
>     def mi_metodo(self):
>         return self.param1
> ```
> • `class`: declara el molde.  
> • `__init__`: constructor (se ejecuta automáticamente al crear un objeto).  
> • `self`: referencia al **objeto actual**. Siempre va como primer parámetro en métodos.


> | Concepto | ¿Dónde se define? | ¿Se comparte entre objetos? | Uso típico |
> |---|---|---|---|
> | **Atributo de clase** | Fuera de `__init__` | Sí, por todas las instancias | Constantes, contadores globales |
> | **Atributo de instancia** | Dentro de `__init__` (`self.x`) |  No, es único por objeto | Datos propios de cada objeto |
> | **Método de instancia** | `def metodo(self):` |  Depende del estado del objeto | Comportamiento específico |
> | **Herencia** | `class Hijo(Padre):` |  Hereda atributos/métodos | Reutilizar código sin repetirlo |

> • `self` **NO es palabra reservada**, es convención. Podrías llamarlo `this`, pero no es lo común  
> • Los métodos **siempre** deben recibir `self` como primer argumento.  
`  
> • Una clase no ejecuta código al definirla; solo se ejecuta al crear instancias: `obj = MiClase()`

````python
#  Clase base con atributos de clase e instancia
class Perro:
    especie = "Canis lupus"  # Atributo de clase (compartido)
    contador = 0                        # Atributo de clase para contar instancias

    def __init__(self, nombre, edad):
        self.nombre = nombre            # Atributo de instancia (único)
        self.edad = edad
        Perro.contador += 1             # Incrementamos contador global

    # Método de instancia
    def presentar(self):
        return f"Soy {self.nombre}, tengo {self.edad} años y soy un {self.especie}"

    def ladrar(self, veces=1):
        return "¡Guau! " * veces

# Creo instancias
firulais = Perro("Firulais", 3)
luna = Perro("Luna", 5)


# Uso mis métdos
print(" Instancias creadas:", Perro.contador)
print("\n", firulais.presentar())
print( luna.ladrar(2))

````

````python
# Diferencias de atributos
print("\n ¿Comparten especie?", firulais.especie == luna.especie)  # True
print(" ¿Comparten nombre?", firulais.nombre == luna.nombre)      # False
````

# Abstracción y Encapsulamiento

-  Abstracción：Se centra en *qué* debe hacer el objeto， ocultando *cómo* lo hace internamente. Define una interfaz clara que las clases hijas están obligadas a cumplir

-  Encapsulamiento：Agrupa datos y comportamientos， restringiendo el acceso directo a atributos críticos para evitar estados inconsistentes

| Principio | ¿Dónde se nota？ | Herramienta en Python | Resultado |
|---|---|---|---|
| Abstracción | Diseño de contratos / clases base | `from abc import ABC, abstractmethod` | Código obligatorio y predecible |
| Encapsulamiento | Manejo de atributos internos | `__nombre` + `get_x()` / `set_x()` | Datos seguros y validados antes de asignar |


````python
from abc import ABC, abstractmethod

# ABSTRACCIÓN：Define el contrato (qué debe hacer cualquier dispositivo)
class Dispositivo(ABC):
    @abstractmethod
    def encender(self): pass

    @abstractmethod
    def obtener_estado(self): pass

# ENCAPSULAMIENTO
class Smartphone(Dispositivo):
    def __init__(self, marca, bateria_inicial=50):
        self.marca = marca
        self.__bateria = bateria_inicial      # Privado (encapsulado)
        self.__encendido = False

    # Implementación obligatoria por abstracción
    def encender(self):
        if not self.__encendido:
            self.__encendido = True
            print(f" {self.marca} encendido")

    def obtener_estado(self):
        if not self.__encendido:
            return " Apagado"
        return f"Batería: {self.__bateria}%"

    # SETTER con validación (encapsulamiento en acción)
    def cargar(self, porcentaje):
        if not self.__encendido:
            print(" Enciende el dispositivo primero")
            return
        if 0 <= porcentaje <= 100:
            self.__bateria = porcentaje
            print(f" Cargado al {porcentaje}%")
        else:
            print(" Valor inválido (0-100)")

# Prueba de ejecución
tel = Smartphone("NovaX", 70)
print(tel.obtener_estado())
tel.encender()               # Método abstracto implementado
tel.cargar(85)
print(" Estado final:", tel.obtener_estado())
````

Métodos Constructores, de Acceso y Modificadores

-  Constructor (`__init__`): Método especial que se ejecuta automáticamente al instanciar un objeto. Inicializa sus atributos.
-  Método de Acceso (Getter): Permite **leer** el valor de un atributo (ideal si está encapsulado con `__`).
- Método Modificador (Setter): Permite **modificar** un atributo, generalmente con validaciones para mantener la integridad de los datos.

#
| Concepto | Firma típica | ¿Para qué sirve? | Buena práctica |
|---|---|---|---|
| Constructor | `def __init__(self, ...):` | Inicializar estado único del objeto | Usar parámetros con valores por defecto opcionales |
| Getter | `def get_atributo(self):` | Obtener valor de forma segura | Evita acceso directo a atributos privados |
| Setter | `def set_atributo(self, valor):` | Modificar con validación | Rechaza datos inválidos antes de asignar |

````python
class CuentaBancaria:
    # Constructor: inicializa el objeto al crearlo
    def __init__(self, titular, saldo_inicial=0.0):
        self.titular = titular          # Atributo público
        self.__saldo = saldo_inicial    # Atributo privado (encapsulado con __)
        print(f"Cuenta creada para: {self.titular}")

    # Método de Acceso (Getter)
    def get_saldo(self):
        return self.__saldo

    # Método Modificador (Setter) con validación
    def set_saldo(self, nuevo_saldo):
        if nuevo_saldo < 0:
            print("Error: El saldo no puede ser negativo.")
            return False
        self.__saldo = nuevo_saldo
        print(f"Saldo modificado correctamente a: ${self.__saldo:.2f}")
        return True

    # Método de instancia adicional
    def depositar(self, monto):
        if monto > 0:
            self.__saldo += monto
            print(f"Depósito de ${monto:.2f} realizado.")
        else:
            print("El monto debe ser mayor a 0.")


if __name__ == "__main__":
    # Crear una cuenta
    cuenta = CuentaBancaria("Ana García", 50.00)

    # Consultar saldo inicial
    print(f"\n Saldo actual: ${cuenta.get_saldo():}")

    # 3. Realizar un depósito válido
    cuenta.depositar(100.0)

    # Intentar depósito inválido
    cuenta.depositar(-20.0)

    # Modificar saldo con el setter
    cuenta.set_saldo(200.0)

    # Intentar poner saldo negativo
    cuenta.set_saldo(-10.0)

    # Saldo final
    print(f"\nSaldo final: ${cuenta.get_saldo():.2f}")
````

¿Para qué sirven estos conceptos?

- Constructor（`__init__`）：Es como el "formulario de registro" que llenas al crear una cuenta. Define qué datos mínimos necesita el objeto para existir
  Por ejemplo, rear un usuario, una cuenta bancaria, un producto en un carrito de compras

- Encapsulamiento（`__atributo` + getters/setters）：Es como una caja fuerte：puedes guardar y consultar el contenido，pero no manipularlo directamente sin permiso.
  Por ejemplo, evitar que un saldo bancario se vuelva negativo，validar que una edad sea real，proteger contraseñas.

- Abstracción（clases abstractas）：Es como un "contrato" o "receta base"：define qué pasos deben seguirse，pero cada quien los cocina a su estilo.
  Por ejemplo, sistemas de pago（todos deben tener `procesar_pago()`pero cada uno lo hace diferente
