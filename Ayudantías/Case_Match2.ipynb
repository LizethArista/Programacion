###`match/case` – Patrones Secuenciales y Captura de Valores

**Evaluación de arriba hacia abajo**: Python prueba cada `case` en orden. En cuanto uno coincide, ejecuta su bloque y **sale automáticamente** (no existe `break`).  
**Captura de valores**: `case x:`, `case [a, b, *resto]:`, `case {"clave": v}:` extraen partes de la estructura y las asignan a variables.  
**(`if` dentro de case)**: `case x if condicion:` permite refinar la coincidencia sin salir de la estructura.  
** `_`**: Captura todo lo que no coincidió antes. Es obligatorio si quieres evitar errores silenciosos.  
**El orden importa**: Colocar un patrón muy general arriba puede "robar" coincidencias que deberían ir a casos más específicos abajo.



```python
dato_1 = [10,20,30,40]


# CAPTURA EN LISTAS Y SECUENCIAS
# Python recorre los cases de arriba a abajo. El primero que encaja, gana.
print("Desestructuración de listas")
match dato_1:
    case []:
        print("Lista vacía")
    case [unico]:
        print(f"Lista de 1 elemento. Capturado como 'unico' = {unico}")
    case [primero, segundo]:
        print(f"Lista de 2 elementos. Capturados: {primero}, {segundo}")
    case [primero, segundo, *resto]:
        # Captura parcial: primero y segundo se asignan, el resto va a una lista
        print(f"Lista larga. Primero: {primero}, Segundo: {segundo}, Resto: {resto}")
    case _:
        print("No es una lista reconocida")


```

```python
dato_2 = (5, 10)

# (if dentro de case)
# El patrón coincide primero, luego se evalúa la condición. Si falla, salta al siguiente

print("\n Tuplas con guardia condicional")
match dato_2:
    case ("origen", y):
        print(f" Caso especial 'origen'. Y capturado = {y}")
    case (x, y) if x == y:
        print(f"Coordenada en diagonal. x={x}, y={y}")
    case (x, y) if y > 10:
        # Aquí se captura x e y, pero SOLO se ejecuta si y > 10
        print(f"Condición con y (y>10). Capturados: x={x}, y={y}")
    case (x, y):
        print(f"Tupla genérica (2 valores). Capturados: {x}, {y}")
    case _:
        print(" Estructura no reconocida")
```

```python
dato_4 = 42

# ¿Qué pasa si colocas mal los casos?

print("\n Demostración de orden secuencial")
match dato_4:
    case x if x < 0:
        print("Negativo capturado")

    case x if x % 7 == 0:
        # NUNCA se ejecutará si el caso anterior ya capturó el valor.
        print(f" Divisible entre 7. Capturado x = {x}")

    case x if 0 <= x < 50:
        # Este caso se evaluará ANTES que el de "par" o "divisible por 7"
        # Si dato_4 = 42, coincide aquí y SALTA los casos siguientes.
        print(f" Rango bajo (0-49). Capturado x = {x}")

    case x if x % 2 == 0:
        # Igual: solo se evalúa si los anteriores fallaron.
        print(f" Número par. Capturado x = {x}")
    case _:
        print(" No coincide con ningún rango o patrón definido")
```

### El Operador `*` en `match/case`: Captura de "El Resto"

En Python, puedes usar `*` dentro de un patrón de secuencia para capturar **todos los elementos restantes** en una variable tipo lista. Es ideal para procesar comandos, rutas o datos de longitud variable de forma declarativa.

- `case [a, b, *resto]:`  Captura los 2 primeros y pone el resto en `resto` (puede ser `[]`).  
-  `case [cabecera, *cola]:` Separa un "encabezado" del "cuerpo" de datos.  
-  `case [primero, *medio, ultimo]:`  `*` en medio es válido, pero la secuencia debe tener ≥ 2 elementos.  
-  `*` **siempre devuelve una lista**, incluso si no hay elementos restantes.  
-  El flujo sigue siendo secuencial: el primer `case` que encaja gana y sale automáticamente.



```python

# ["AYUDA"]
# ["CARGAR", "datos.csv"]
# ["CARGAR", "a.txt", "b.csv", "c.json"]
# ["FILTRO", ">5", "<10"]
# ["FILTRO", ">5", "<10", "!=0", "&&", "true"]
# ["PROCESAR", 10, 20, 30, 40]
entrada = ["CARGAR", "archivo1.txt", "archivo2.csv", "archivo3.json"]
entrada2 = ["PROCESAR", 10, 20, 30, 40]
entrada3 = ["FILTRO", ">5", "<10", "!=0", "&&", "true"]

match entrada2:
    # Comando sin argumentos
    case ["AYUDA"]:
        print(" Mostrando menú de ayuda general...")

    #  Captura del primer elemento + el RESTO
    # Coincide con listas de longitud >= 2
    case ["CARGAR", principal, *restantes]:
        print(f" Comando: CARGAR")
        print(f"Archivo principal : {principal}")
        print(f"Archivos adicionales: {restantes}")
        print(f"Total a procesar  : {1 + len(restantes)}")

    # Guardia + *: Valida cantidad de argumentos antes de ejecutar
    case ["FILTRO", criterio, *extras] if len(extras) <= 2:
        print(f" Comando: FILTRO")
        print(f"  Criterio base   : {criterio}")
        print(f"  Condiciones extra: {extras}")
        print(" Validación: <= 2 condiciones extra permitidas.")

    case ["FILTRO", *_]:
        print(" Filtro rechazado: Demasiadas condiciones o estructura inválida.")

    #  `*` en el MEDIO del patrón
    # Captura todo lo que esté entre el primer y último elemento
    case [accion, *intermedios, objetivo]:
        print(f" Comando: {accion}")
        print(f"  Parámetros intermedios: {intermedios}")
        print(f"  Objetivo/Resultado esperado: {objetivo}")

    # Caso genérico para cualquier otra lista
    case [comando, *argumentos]:
        print(f"Comando no registrado: '{comando}'")
        print(f" Argumentos recibidos : {argumentos}")

    # Fallback final: no es lista o está vacío
    case _:
        print(" Entrada no reconocida")

#  *resto SIEMPRE es una lista []. Si no hay elementos, devuelve []
#  [a, b, *c] requiere longitud mínima de 2.
#  [a, *b, c] requiere longitud mínima de 2.
#  El orden importa: si colocas [_, *_] arriba, "robará" todas las listas.
```
