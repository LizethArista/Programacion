## EJERCICIO 1: Ecuación Cuadrática (Discriminante Positivo)

Diseña un programa que solicite al usuario los coeficientes a, b y c de una ecuación cuadrática de la forma:
  $ ax^2+bx+c=0 $

El programa deberá:

* Validar que a≠0(de lo contrario, no es una ecuación cuadrática)
* Calcular el discriminante: D=b^2-4ac
* Si D>0, calcular y mostrar las dos raíces reales usando: $ x_1=(-b+√D)/2a,x_2=(-b-√D)/2a $
* Si D≤0, mostrar un mensaje indicando que el ejercicio solo resuelve casos con discriminante positivo

**NOTAS:** Antes de iniciar tu codigo debes importar la librería math (línea de cógigo: import math)

```python
 #Solución
 # Ecuación cuadrática

 # Solo se resuelven casos con discriminante positivo (D > 0)
import math

# Pido al usuario los coeficientes
a = float(input("Ingrese el coeficiente a: "))
b = float(input("Ingrese el coeficiente b: "))
c = float(input("Ingrese el coeficiente c: "))

# Valido que sea una ecuación cuadrática
if a == 0:
    print("Error: El coeficiente 'a' no puede ser cero en una ecuación cuadrática.")
else:
    # Calcular discriminante
    discriminante = b**2 - 4*a*c
    print(f"\nEcuación: {a}x^2 + {b}x + {c} = 0")
    print(f"Discriminante (D): {discriminante}")

    if discriminante > 0:
        raiz1 = (-b + math.sqrt(discriminante)) / (2*a)
        raiz2 = (-b - math.sqrt(discriminante)) / (2*a)
        print(f"Las soluciones son: ")
        print(f"Raíz 1: {raiz1}")
        print(f"Raíz 2: {raiz2}")
    else:
        print("El discriminante no es positivo. Este programa solo resuelve casos con D > 0")
```

## EJERCICIO 2: Registro y Procesamiento de Datos de Ventas

Una tienda requiere registrar la venta de un producto. Diseñe un programa que solicite la siguiente información:

1.	Nombre del vendedor
2.	Clave del producto
3.	Cantidad vendida (entero positivo)
4.	Precio unitario (flotante positivo)
5.	Tipo de cliente ('nuevo' o 'frecuente')

El programa deberá:

*	Calcular el subtotal: cantidad * precio_unitario

*	Aplicar un descuento del 10% si el cliente es 'frecuente'

*	Calcular el total a pagar

*	Mostrar un resumen con los datos ingresados y los cálculos realizados


```python
#Solución

# Solicito mis datos
nombre = input("Nombre del vendedor: ")
clave = input("Clave del producto: ")
cantidad = int(input("Cantidad vendida: "))
precio = float(input("Precio unitario: "))
tipo = (input("Tipo de cliente (nuevo/frecuente): ")).lower()

# Calculos...
if cantidad > 0 and precio > 0:
    subtotal = cantidad * precio

    if tipo == "frecuente":
        descuento = subtotal * 0.10
    else:
        descuento = 0

    total = subtotal - descuento

    # 3. SALIDAS
    print("\n--- RESUMEN DE COMPRA ---")
    print("Vendedor:", nombre)
    print("Subtotal: $", subtotal)
    print("Descuento:", descuento)
    print("TOTAL: $", total)

else:
    print("Error: La cantidad y el precio deben ser positivos.")
```
