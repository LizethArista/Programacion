```python
comando = input("Por ejemplo actualizar 42  o  buscar python java: ").split()

match comando:
    case ["actualizar", id]:
        print(f" Actualizando ID: {id}")
    case ["buscar", *terminos]:
        print(f" Buscando: {terminos}")
    case _:
        print("No reconocido")
```

```python
comando = input("Ej: mover 5  o  girar izq: ").split()

match comando:
    case ["mover", pasos] if pasos.isdigit():
        print(f"Avanzando {pasos} pasos")
    case ["girar", direccion] if direccion in ["izq", "der"]:
        print(f"Girando a la {direccion}")
    case _:
        print("Ingresa un comando o dato válido")
```

```python
# Ignora todo después del ID
comando = ["actualizar", "usuario", 42, "campo1", "campo2"]

match comando:
    case ["actualizar", _, id, *_]:
        print(f"Actualizar {id}")

# Captura el resto en una variable
comando = ["buscar", "python", "java", "rust"]

match comando:
    case ["buscar", *terminos]:
        print(terminos)
# Ignora el resto con *_
comando = ["comando", "arg1", "arg2", "arg3", "arg4"]

match comando:
    case ["comando", primero, *_]:
        print(primero)
```

## Recodemos...

En Python, **todo es un objeto**. Una **clase** es el "plano" o molde, y un **objeto** es una instancia concreta de ese plano.

Cada objeto tiene:
- **Atributos**: datos o propiedades que lo describen.
- **Métodos**: funciones que definen su comportamiento.


## Composición de Objetos

La **composición** es un principio de diseño donde un objeto **contiene a otro(s)** como parte de su estructura. Se conoce como una relación **"tiene un"** (`has-a`).

Ejemplo: Un `Auto` **tiene un** `Motor`, **tiene un** `SistemaDeFrenos`, etc.

- Permite cambiar componentes en tiempo de ejecución
- Fomenta la reutilización de código

```python
# Clase independiente: Motor
class Motor:
    def __init__(self, tipo, potencia):
        self.tipo = tipo
        self.potencia = potencia

    def encender(self):
        return f"Motor {self.tipo} encendido ({self.potencia} CV)"

# Clase que NECESITA O UTILIZA al Motor
class Auto:
    def __init__(self, marca, motor):
        self.marca = marca
        self.motor = motor  # Auto "tiene un" Motor

    def arrancar(self):
        # El auto delega el trabajo al motor
        return f"{self.marca}: {self.motor.encender()}"

mi_motor = Motor("V6", 320)
mi_auto = Auto("Toyota", mi_motor)

print(mi_auto.arrancar())
```

```python
# CLASES COMPONENTES
class CPU:
    def __init__(self, modelo, nucleos):
        self.modelo = modelo
        self.nucleos = nucleos

    def procesar(self, tarea):
        return f" CPU {self.modelo} ({self.nucleos} núcleos) procesando: {tarea}"

class RAM:
    def __init__(self, gb):
        self.gb = gb

    def cargar(self):
        return f"RAM: {self.gb} GB disponibles"

class Disco:
    def __init__(self, tipo, tb):
        self.tipo = tipo
        self.tb = tb

    def guardar(self, archivo):
        return f"Disco {self.tipo} ({self.tb} TB): guardando {archivo}"


# CLASE CONTENEDORA (usa los objetos como atributos)
class Computadora:
    def __init__(self, nombre, cpu, ram, disco):
        self.nombre = nombre
        self.cpu = cpu
        self.ram = ram
        self.disco = disco

    def boot(self):
        print(f"Iniciando {self.nombre}...")

        # la computadora le pide a sus objetos internos que trabajen
        print(self.ram.cargar())
        print(self.disco.guardar("sistema.bin"))
        print(self.cpu.procesar("arranque"))
```

```python
cpu_rapida = CPU("Intel i7", 8)
memoria = RAM(16)
almacenamiento = Disco("SSD", 1)

# Creamos nuestro objeto computadora, teniendo a mis componentes como atributos
mi_pc = Computadora("Workstation Pro", cpu_rapida, memoria, almacenamiento)

mi_pc.boot()

print("\n Tipo de los atributos:")
print(f"self.cpu  -> {type(mi_pc.cpu)}")
print(f"self.ram  -> {type(mi_pc.ram)}")
print(f"self.disco-> {type(mi_pc.disco)}")
```
