# Ciclo `for`
Veamos `for`, es una herramienta de Python para iterar sobre colecciones de datos

### 1. ¿Qué es un bucle `for`?

El bucle `for` se utiliza para **iterar** (recorrer) una secuencia de elementos. A diferencia del `while`, el `for` sabe exactamente cuántas veces se va a ejecutar.

#### Sintaxis básica:
```python
for elemento in secuencia:
    # Código que se repite
    print(elemento)

````python
frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print(fruta)

# La ventaja del ciclo for es que no se necesita usar un contador
#manual ni usar bucles infinitos
````

````python
#  Mi Primera Lista

# Crea una lista con 5 números favoritos
# Usa un bucle for para imprimir cada uno multiplicado por 2

numeros_favoritos = [5, 10, 15, 20, 25]

for numero in numeros_favoritos:
    print(f"{numero} x 2 = {numero * 2}")
````

## 2. La función `range()`

`range()` genera una secuencia de números. Es muy útil cuando quieres repetir algo un número específico de veces.

### Formas de usar `range()`:
- `range(5)` → 0, 1, 2, 3, 4
- `range(2, 6)` → 2, 3, 4, 5
- `range(0, 10, 2)` → 0, 2, 4, 6, 8 (de 2 en 2)

```python
for i in range(5):
    print(i)  # Imprime 0 al 4

````python
for i in range(5):
    print(i)  # Imprime 0 al 4
````

````python
#  Tabla de Multiplicar

# Pide al usuario un número y muestra su tabla de multiplicar del 1 al 10

numero = int(input("Ingresa un número para ver su tabla: "))

print(f"\n--- Tabla del {numero} ---")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
````

## Ejercicio 3 (Versión `for`)

En la clase anterior, recorrimos un texto usando `while` y un índice manual. Ahora haremos lo mismo con `for`.

### ¿Cuál es la diferencia?
- **Con `while`:** Necesitas crear una variable `indice`, inicializarla en 0, verificar que sea menor al largo del texto y aumentarla manualmente en cada vuelta

- **Con `for`:** Python se encarga de recorrer el texto automáticamente. Es más limpio y menos propenso a errores


```python
def contar_vocales(texto):
    vocales = "aeiouAEIOU"
    contador = 0
    indice = 0

    while indice < len(texto):
        if texto[indice] in vocales:
            contador += 1
        indice += 1
        
    return contador
contar_vocales("hOLA")

````python
# Ejemplo de la clase pasda pero ahora usando for....

def contar_vocales_for(texto):
    vocales = "aeiouAEIOU"
    contador = 0

    for letra in texto:
        if letra in vocales:
            contador += 1

    return contador

contar_vocales_for( "programacion")
````

````python

# Crea una función que cuente cuántas letras mayúsculas hay en un texto

def contar_mayusculas(texto):
    contador = 0

    for letra in texto:
        if letra.isupper():
            contador += 1
    return contador

assert contar_mayusculas("Hola Mundo") == 2
assert contar_mayusculas("python") == 0
assert contar_mayusculas("HTML y CSS") == 7
````

## 4. Ciclos Anidados

Un bucle `for` dentro de otro bucle `for`

Es util para trabajar con tablas o matrices

```python
for i in range(3):      # Bucle externo
    for j in range(3):  # Bucle interno
        print(f"({i}, {j})", end=" ")
    print()  # Salto de línea

#### Triángulo de Estrellas

Crea un patrón de estrellas que crece en cada fila

Ejemplo para 5 filas:
# *
# **
# ***
# ****
# *****


````python
filas = 5

for i in range(1, filas+1 ):
    for j in range(i):
        print("*", end="") # end= es un parámetro opcional que indica que
                            #  se imprime al final de LA MISMA linea
    print()  # Salto de línea al final de cada fila

````

## 5. Funciones Útiles con `for`

### `enumerate()`
Obtienes el índice y el valor al mismo tiempo.
```python
for indice, valor in enumerate(["a", "b", "c"]):
    print(f"Índice {indice}: {valor}")

````python
for indice, valor in enumerate(["a", "b", "c"]):
    print(f"Índice {indice}: {valor}")
````

````python
nombres = ["Ana", "Luis"]
edades = [25, 30]
for nombre, edad in zip(nombres, edades):
    print(f"{nombre} tiene {edad} años")
````

````python
# Lista de Compras Numerada

# Usa enumerate para mostrar una lista de compras con números
# Luego usa zip para mostrar precios junto a los productos

productos = ["Manzanas", "Pan", "Leche", "Huevos"]
precios = [2.5, 1.2, 0.9, 3.0]

print("--- LISTA DE COMPRAS ---")
for indice, producto in enumerate(productos, start=1):
    print(f"{indice}. {producto}")

print("\n--- PRECIOS ---")
for producto, precio in zip(productos, precios):
    print(f"{producto}: ${precio}")
````
