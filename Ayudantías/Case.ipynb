### `match/case` vs `if/elif/else`: ¿En qué se parecen y en qué difieren?

| Característica | `if / elif / else` | `match / case` |
|---|---|---|
| **Qué evalúa** | Cualquier expresión booleana (`x > 5`, `len(lista) == 3`, `función()`) | Un **valor o estructura específica** contra patrones definidos |
| **Legibilidad** | Se alarga rápido con muchas categorías discretas | Más limpio y visual para clasificaciones fijas |



- Usa `if/elif` cuando necesites evaluar **condiciones lógicas, rangos o cálculos**.  
- Usa `match/case` cuando quieras **clasificar o categorizar ** un valor conocido  



```python
# Clasificación básica de nivel de riesgo
# En actuaría, clasificamos riesgos para asignar factores de ajuste a la prima

nivel_riesgo = "Alto"

match nivel_riesgo:
    case "Bajo" | "muy bajo ":
        factor_ajuste = 0.85  # Descuento por baja probabilidad de siniestro
        print(" Riesgo Bajo: Factor de descuento aplicado (0.85)")
    case "Medio":
        factor_ajuste = 1.00  # Tarifa estándar
        print(" Riesgo Medio: Tarifa base sin ajustes (1.00)")
    case "Alto":
        factor_ajuste = 1.25  # Recargo por mayor frecuencia histórica
        print(" Riesgo Alto: Recargo aplicado (1.25)")
    case "Crítico":
        factor_ajuste = 1.50  # Requiere evaluación técnica especial
        print(" Riesgo Crítico: Requiere revisión de suscripción")
    case _:
        # aptura valores no contemplados
        factor_ajuste = 1.00
        print(" Riesgo no clasificado: Se aplica tarifa neutral")

#print(f" Factor de riesgo final: {factor_ajuste}")
```

### ¿Cómo funciona `match/case`?
- `match variable:` inicia la evaluación.
- Python recorre los `case` **de arriba hacia abajo**. En cuanto encuentra una coincidencia, ejecuta ese bloque y **sale automáticamente**.
- `|` permite agrupar valores equivalentes (ej: `case "Bajo" | "MuyBajo":`).
- `case _:` captura cualquier valor no contemplado (obligatorio en modelos actuariales para evitar errores silenciosos).
-  **No se usa `break`**

```python
# Agrupación por tipo de póliza
# Diferentes seguros tienen distintas distribuciones de probabilidad base

tipo_poliza = input("Tipo de póliza:")

match tipo_poliza:
    case "Auto" | "Moto":
        prob_siniestro = 0.12  # 12% probabilidad anual (frecuencia histórica)
        print("Póliza Vehicular: Prob. siniestro = 12%")
    case "Vida":
        prob_siniestro = 0.008  # 0.8% (tablas de mortalidad)
        print("Póliza de Vida: Prob. siniestro = 0.8%")
    case "Hogar" | "Negocio":
        prob_siniestro = 0.05   # 5% (incendios, robos, desastres naturales)
        print("Póliza Patrimonial: Prob. siniestro = 5%")
    case "Salud":
        prob_siniestro = 0.35   # 35% (alto uso por consultas frecuentes)
        print("Póliza de Salud: Prob. siniestro = 35%")
    case _:
        prob_siniestro = 0.10
        print("Sin clasificacion: Se usa probabilidad neutral (10%)")

```

### Cuando la probabilidad depende de condiciones extra
En actuaría, no solo importa el tipo de seguro, sino también **edad, historial o hábitos**. Python permite añadir condiciones dentro del `case` usando `if`.

- Sintaxis: `case patron if condicion:`
Se evalúa primero el patrón, y solo si coincide, se verifica la condición. Si no se cumple, Python salta al siguiente `case`.

```python
# Ajuste por edad usando Guard Clauses (if dentro de case)
# Simulamos un clasificador que modifica la prima según el grupo de edad

edad = int(input("Ingresa tu edad: "))

match edad:
    case e if e < 18:
        ajuste_edad = 1.10
        print("Menor de edad: +10% por falta de historial estadístico")
    case e if 18 <= e <= 30:
        ajuste_edad = 0.95
        print("Joven adulto: -5% por mejor perfil probabilístico")
    case e if 31 <= e <= 55:
        ajuste_edad = 1.00
        print("Edad madura: Tarifa actuarial estándar")
    case e if 56 <= e <= 70:
        ajuste_edad = 1.20
        print("Adulto mayor: +20% por aumento de siniestralidad")
    case _:
        ajuste_edad = 1.50
        print("Edad fuera de rango: Requiere suscripción especial")

print(f"Factor de ajuste por edad: {ajuste_edad}")
```



-  Usa `match` para **categorías claras** y `if/elif` para **límites continuos o fórmulas complejas**

```python
# Calculadora de Prima Base
# Unimos todo: tipo de póliza + nivel de riesgo + ajuste por edad

# Tareita: Juega con los valores para ver distintos escenarios :)
tipo = "Vida"
riesgo = "Alto"
edad_cliente = 60

# Probabilidad base por tipo
match tipo:
    case "Auto" | "Moto": prob_base = 0.12
    case "Vida": prob_base = 0.008
    case "Hogar": prob_base = 0.05
    case _: prob_base = 0.10

# Factor por nivel de riesgo
match riesgo:
    case "Bajo": factor_riesgo = 0.85
    case "Medio": factor_riesgo = 1.00
    case "Alto": factor_riesgo = 1.25
    case _: factor_riesgo = 1.00

# Factor por edad
match edad_cliente:
    case e if e < 25: factor_edad = 1.15
    case e if 25 <= e <= 50: factor_edad = 1.00
    case e if 51 <= e <= 70: factor_edad = 1.20
    case _: factor_edad = 1.40

# Cálculo final
prima_base = 500
prima_final = prima_base * factor_riesgo * factor_edad


print(f"Tipo de póliza : {tipo} | Prob. base: {prob_base:.3f}")
print(f"Nivel de riesgo: {riesgo} | Factor: {factor_riesgo}")
print(f"Edad del cliente: {edad_cliente} años | Factor: {factor_edad}")
print(f"PRIMA FINAL ESTIMADA: ${prima_final:,.2f} ")
```
