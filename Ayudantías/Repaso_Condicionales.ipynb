## RECUERDA... ¿Qué es un if?

Imagina que eres un guardia de seguridad:
*   **SI** tienes invitación -> **Entras**.
*   **SI NO** tienes invitación -> **Te quedas fuera**.

En Python, usamos la palabra clave if (si) para esto. Es muy importante la **indentación** (los espacios al inicio de la línea). Todo lo que esté dentro pertenece al if

### Operadores de Comparación
| Operador | Significado | Ejemplo |
| :---: | :--- | :--- |
| `==` | Igual a | `5 == 5` (True) |
| `!=` | Diferente de | `5 != 3` (True) |
| `>` | Mayor que | `10 > 5` (True) |
| `<` | Menor que | `3 < 1` (False) |
| `>=` | Mayor o igual | `5 >= 5` (True) |
| `<=` | Menor o igual | `4 <= 5` (True) |

## Qué pasa si la condición no se cumple? Usamos else (si no). Solo se ejecuta si el if fue falso

**Estructura:**
```python
if condicion:
    # Se ejecuta si es VERDADERO
else:
    # Se ejecuta si es FALSO

````python
# Ejemplo para recordar...
contrasena_usuario = "python123"
contrasena_correcta = "python123"

if contrasena_usuario == contrasena_correcta:
    print("Contraseña correcta, bienvenido")
else:
    print("Contraseña incorrecta")
````

````python
#Ejercicios... (clase pasada)

#Pida un número y verifique si está entre 1 y 100, incluyéndolos en el rango

numero = int(input("Ingresa un número: "))

# Verifico si está en el rango [1, 100]
# Usamos 'and' porque AMBAS condiciones deben ser verdaderas
if numero >= 1 and numero <= 100:
    print(f"El número {numero} está entre 1 y 100")
else:
    print(f"El número {numero} NO está entre 1 y 100")
````

## Múltiples opciones: elif

A veces hay más de dos caminos. Usamos elif (else if) para verificar varias condiciones en orden. Python se detiene en la primera que sea verdadera

````python
# Calificación de estudiantes
# If - Elif - Else
nota = 85

print(f"Tu nota es: {nota}")

if nota >= 90:
    print("Excelente")
elif nota >= 70:
    print("Aprobado")
elif nota >= 60:
    print("Necesitas mejorar")
else:
    print("Reprobado")
````

## 4. Operadores Lógicos (`and`, `or`, `not`)

Podemos combinar condiciones:
*   `and`: Ambas deben ser verdaderas.
*   `or`: Al menos una debe ser verdadera.
*   `not`: Invierte el valor (True -> False).
