# Tuplas
**Objetivo:** Comprender qué son las tuplas, cómo se crean, sus características principales, operaciones disponibles y cuándo utilizarlas en comparación con las listas.

```python
# Una tupla es una colección ORDENADA e INMUTABLE de elementos.
# Se define con paréntesis () o simplemente separando con comas.

# Creación
tupla1 = (1, 2, 3)
tupla2 = "a", "b","c"          # Empaquetado automático (sin paréntesis)
tupla3 = tuple([4, 5, 6])       # Conversión desde lista
tupla_vacia = ()
tupla_un_elemento = (42,)       # La coma es OBLIGATORIA para un solo elemento

print("tupla1:", tupla1)
print("tupla2:", tupla2)
print("tupla3:", tupla3)
print("Tipo tupla_un_elemento:", type(tupla_un_elemento))
```


A diferencia de las listas, **no puedes modificar, agregar ni eliminar elementos** de una tupla una vez creada. Esto las hace:
- Más rápidas en memoria
- Seguras para datos constantes
- Válidas como **claves en diccionarios** o elementos en `set`

```python
# INMUTABILIDAD Y ACCESO A DATOS
mi_tupla = (10, 20, 30, 40, 50)

# Acceso por índice y slicing (igual que listas)
print("Primer elemento:", mi_tupla[0])
print("Último elemento:", mi_tupla[-1])
print("slicing:", mi_tupla[1:4])

#mi_tupla[0] = 99        # TypeError: 'tuple' object does not support item assignment
#mi_tupla.append(60)     # AttributeError: 'tuple' object has no attribute 'append'

# Si deseamos modificar una tupla, podemos convertirla a lista y volverla a convertir a tupla
lista_temp = list(mi_tupla)
lista_temp[0] = 99
mi_tupla = tuple(lista_temp)
print("Tupla modificada:", mi_tupla)
```

## Operaciones y Métodos
Las tuplas no tienen métodos como `.append()` o `.pop()`, pero sí soportan:
- Concatenación `+` y repetición `*`
- Operadores de pertenencia `in` / `not in`
- Solo dos métodos nativos: `.count()` y `.index()`
- Funciones integradas: `len()`, `min()`, `max()`, `sum()`

```python
# OPERACIONES CON TUPLAS

t1 = (1, 2)
t2 = (3, 4)

# Concatenación y repetición (crean NUEVAS tuplas, no modifican las originales)
t_concat = t1 + t2
t_rep = t1 * 3
print("Concatenada:", t_concat)
print("Repetida:", t_rep)

```

```python
datos = (10, 20, 10, 30, 10, 40)
print("Cantidad de 10:", datos.count(10))       # Cuántas veces aparece
print("Índice de 30:", datos.index(30))         # Primera posición encontrada

```

```python
# Otras funciones...
print("Longitud:", len(datos))
print("Mínimo:", min(datos))
print("Máximo:", max(datos))
print("Suma:", sum(datos))
print("¿Está el 20?:", 20 in datos)
print("¿Está el 99?:", 99 in datos)
```

Podemos extraer los valores de una tupla en variables individuales de forma limpia y segura.

```python
# extraemos los elementos individuales
coordenadas = (4.5, 8.2)
x, y = coordenadas
print(f"x: {x}, y: {y}")

```

```python
# Intercambiamos variables
a, b = 10, 20
a, b = b, a
print(f"Intercambio: a={a}, b={b}")


```

```python

info = ("Ana", 25, "Ing.", "Madrid", "Python")
nombre, edad, *resto = info
print(f"Nombre: {nombre}, Edad: {edad}, Resto: {resto}")

primer, *medio, ultimo = (1, 2, 3, 4, 5, 6)
print(f"Primer: {primer}, Medio: {medio}, Último: {ultimo}")
```

```python
# Podemos retornar múltiples valores desde una función
def dividir_y_modulo(a, b):
    return a // b, a % b  # se tiene una tupla

cociente, residuo = dividir_y_modulo(17, 5)
print(f"17 ÷ 5 = {cociente} con residuo {residuo}")
```
