#  Clases, `__init__`, `self` y Atributos

- **Clase**: Es un "molde" o plano para crear objetos. Define qué datos guardan (atributos) y qué pueden hacer (métodos).
- `__init__`: Constructor. Se ejecuta automáticamente al crear un objeto y sirve para inicializar su estado.
- `self`: Referencia al objeto actual. **No es palabra reservada**, es convención. Siempre va como primer parámetro en métodos de instancia.
- **Atributo de Clase**: Se define fuera de `__init__`. Se comparte entre TODAS las instancias (ej. constantes, contadores).
- **Atributo de Instancia**: Se define dentro de `__init__` (`self.x`). Es único para cada objeto creado.

## Distribución Binomial
Modela el **número de éxitos** en `n` ensayos independientes con probabilidad constante `p`.

 - **Fórmula**: $P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$
 - **Esperanza**: $E[X] = np$
 - **Varianza**: $Var(X) = np(1-p)$

*Aplicación en la actuaría*:

Número de siniestros esperados en una cartera fija de `n` pólizas homogéneas, donde cada póliza tiene probabilidad `p` de presentar un siniestro en el periodo.

```python

import math


# DISTRIBUCIÓN BINOMIAL B(n, p)
# Modela el número de "éxitos" en n ensayos independientes con probabilidad constante p.
# En seguros: "éxito" = que ocurra un siniestro.
# n = número de pólizas expuestas al riesgo.
# p = probabilidad individual de siniestro.
# P(X=k) = C(n,k) * p^k * (1-p)^(n-k)
# E[X] = n*p  |  Var(X) = n*p*(1-p)

class DistribucionBinomial:
    # ATRIBUTO DE CLASE: Pertenece a la clase, NO a las instancias.
    # Se comparte entre TODOS los objetos que creemos.
    tipo_distribucion = "Discreta"
    total_modelos_creados = 0

    # CONSTRUCTOR: Se ejecuta automáticamente al crear el objeto.
    # Inicializa los datos propios de CADA póliza/modelo.
    def __init__(self, n, p):
        # ATRIBUTOS DE INSTANCIA: Prefijo self. indica que pertenecen al objeto
        # que se está creando en este momento.
        self.n = n  # Número de ensayos / pólizas en la cartera
        self.p = p  # Probabilidad individual de siniestro

        # Modificamos el atributo de clase para llevar conteo global.
        # Accedemos vía NombreClase.atributo, NO via self, para evitar confusiones.
        DistribucionBinomial.total_modelos_creados += 1

    # MÉTODO DE INSTANCIA: Calcula la probabilidad puntual P(X = k)
    def probabilidad(self, k):
        # Validación estadística: k no puede ser negativo ni mayor que n
        if not (0 <= k <= self.n):
            return 0.0
        # Cálculo combinatorio: math.comb(n, k) = n! / (k! * (n-k)!)
        comb = math.comb(self.n, k)
        # Aplicación directa de la fórmula de la Binomial
        return comb * (self.p ** k) * ((1 - self.p) ** (self.n - k))

    # MÉTODO DE INSTANCIA: Esperanza matemática E[X]
    def esperanza(self):
        # En seguros: número promedio de siniestros esperados en la cartera.
        return self.n * self.p

    # MÉTODO DE INSTANCIA: Varianza Var(X)
    def varianza(self):
        # Mide la dispersión/riesgo alrededor del promedio esperado.
        return self.n * self.p * (1 - self.p)


```

```python
# CREACIÓN DE INSTANCIAS (OBJETOS)

# Al ejecutar esto, Python llama automáticamente a __init__ con los parámetros.
# Se crean DOS objetos independientes
auto_cartera = DistribucionBinomial(n=20, p=0.15)  # 20 pólizas auto, 15% siniestralidad

vida_cartera = DistribucionBinomial(n=100, p=0.02) # 100 pólizas vida, 98% supervivencia
```

```python
# SALIDAS

print(f"Tipo de distribución (compartido entre clases): {vida_cartera.tipo_distribucion}")
print(f"Modelos creados (atributo de clase): {DistribucionBinomial.total_modelos_creados}")
print(f"Pólizas en cartera (atributo de instancia): {auto_cartera.n}")
print(f"Probabilidad individual de siniestro (atributo de instancia): {auto_cartera.p}")
```

```python
# Cada objeto usa SUS PROPIOS valores de self.n y self.p al llamar métodos
print(f"P(Exactamente 3 siniestros en auto) = {auto_cartera.probabilidad(3):}")
print(f"Siniestros esperados E[X] (auto)    = {auto_cartera.esperanza():}")
print(f"Varianza Var(X) (auto)              = {auto_cartera.varianza():}")

print(f"P(6 siniestros en vida)= {vida_cartera.probabilidad(6):.8f}")
print(f"Siniestros esperadosE[X] vida = {vida_cartera.esperanza():}")
print(f"Variaanza Var(X) vida ={vida_cartera.varianza(): } ")

sd = math.sqrt(vida_cartera.varianza())
print(auto_cartera.esperanza()-sd, auto_cartera.esperanza()+sd)
```

```python
print("\n¿Comparten el atributo de CLASE 'tipo_distribucion'?")
# True: ambos apuntan al mismo valor en memoria de la clase
print(f"   {auto_cartera.tipo_distribucion is vida_cartera.tipo_distribucion}")
print

print("\n ¿Comparten el atributo de INSTANCIA 'n'?")
# False: cada objeto tiene su propia copia en memoria
print(f"   {auto_cartera.n is vida_cartera.n}")
```

# Encapsulamiento, Getters y Setters

- **Encapsulamiento**: Agrupa datos y comportamientos, restringiendo el acceso directo a atributos críticos para evitar estados inconsistentes. En Python se usa el prefijo `_atributo` (convención de "protegido").
- **Getter**: Método para **leer** un atributo encapsulado de forma segura.
- **Setter**: Método para **modificar** un atributo, idealmente con validaciones que rechacen datos inválidos antes de asignarlos.

### Póliza y Reserva Técnica
- **Prima**: Precio que paga el asegurado. Debe ser siempre positiva y ajustarse a tarifas técnicas.
- **Reserva Técnica**: Fondo que la aseguradora constituye para cubrir siniestros futuros. Se calcula con fórmulas regulatorias y no debe manipularse arbitrariamente.
- **Estado de póliza**: Activa / Cancelada. Las operaciones solo deben permitirse si está activa.

 *Por qué encapsular*:
  Para evitar que un programador asigne primas negativas o modifique reservas sin validación actuarial, lo que generaría errores de solvencia.

```python


class PolizaSeguro:
    def __init__(self, titular, suma_asegurada, prima_anual):
        # Atributo público: información descriptiva, no crítica para cálculos internos
        self.titular = titular

        # ATRIBUTOS ENCAPSULADOS (convención Python: prefijo _)
        # El _ indica: "No accedas/modifiques directamente. Usa get/set"
        self._suma_asegurada = suma_asegurada
        self._prima_anual = prima_anual
        self._saldo_reserva = 0.0  # Inicia en 0, se calcula con el tiempo
        self._activa = True        # Estado inicial de la póliza

        print(f"Póliza creada para: {self.titular}")

    # GETTER: Método de acceso seguro
    # Devuelve la prima sin permitir que nadie la sobrescriba directamente
    def get_prima(self):
        return self._prima_anual

    # SETTER: Método modificador con VALIDACIÓN ACTUARIAL
    # Rechaza datos inválidos ANTES de asignarlos al objeto
    def set_prima(self, nueva_prima):
        # una prima negativa o cero no tiene sentido actuarial
        if nueva_prima <= 0:
            print("La prima debe ser estrictamente positiva.")
            return False
        #Si pasa la validación, actualizamos el atributo encapsulado
        self._prima_anual = nueva_prima
        print(f"Prima actualizada a: ${nueva_prima:,.2f}")
        return True

    # Cálculo de reserva técnica
    def actualizar_reserva(self, anio_poliza):
        # solo se calcula si la póliza está vigente
        if not self._activa:
            print("La póliza está dada de baja. No se puede calcular la reserva.")
            return

        # Prima × Años × Factor técnico (5% anual)
        # En la realidad usarías tablas regulatorias y tasas técnicas oficiales
        self._saldo_reserva = self._prima_anual * anio_poliza * (1.05 ** anio_poliza)
        print(f"Reserva técnica al año {anio_poliza}: ${self._saldo_reserva:,.2f}")

    # GETTER para la reserva
    def get_reserva(self):
        return self._saldo_reserva
```

```python


# PRUEBA DE ENCAPSULAMIENTO Y VALIDACIONES
poliza = PolizaSeguro("Carlos Méndez", 50000, 800)

# Cálculo inicial de reserva (año 3)
poliza.actualizar_reserva(3)

# Intento de modificación INVÁLIDA (simulando un error de usuario/sistema)
# Si el atributo fuera público (self.prima_anual = -50), el sistema colapsaría.
# Gracias al encapsulamiento + setter, la validación lo rechaza:
print("\n Intentando asignar prima negativa con el setter:")
poliza.set_prima(-50)  # El setter intercepta y rechaza

# Modificación VÁLIDA
print("\n Asignando prima válida:")
poliza.set_prima(950)

# Recalculamos reserva con la nueva prima
poliza.actualizar_reserva(3)

# Acceso final seguro vía getter
print(f"\n Reserva final accesible vía getter: ${poliza.get_reserva():,.2f}")
```
