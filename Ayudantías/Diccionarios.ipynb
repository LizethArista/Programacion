# Diccionarios
### ¿Qué es un diccionario (`dict`)?
- Es una estructura que almacena pares **clave: valor** (`key: value`).
- Las claves deben ser **únicas** e **inmutables** (hasheables).
- Los valores pueden ser de **cualquier tipo** (listas, tuplas, otros diccionarios, funciones, etc.).

```python
# creacion de un diccionario
# Sintaxis: {clave: valor, clave: valor}
estudiante = {
    "nombre": "Ana",
    "edad": 22,
    "carrera": "Ingeniería",
    "notas": (8.5, 9.0, 7.8)  # tenemos un tupla como valor
}

print("Diccionario completo:", estudiante)
print("\n Acceso por clave:", estudiante["nombre"])
print(" Acceso seguro con .get():", estudiante.get("email", "No registrado"))
# .get() evita KeyError si la clave no existe
```

## ¿Cómo funciona internamente? (Tablas Hash)
Python usa una **tabla hash** para almacenar diccionarios:
1. Calcula el `hash()` de la clave.
2. Usa ese hash para determinar la posición en memoria.
3. Por eso las claves **deben ser inmutables** (strings, números, tuplas puras).
4. Si la clave cambiara después de insertarla, su hash cambiaría → el sistema ya no sabría dónde buscar el valor


```python

# algunas opraciones... (Crear, Leer, Actualizar, Eliminar)

config = {"theme": "dark", "lang": "es", "version": 1.2}

# Para agregar o actualizar se usa el mismo operador =
config["theme"] = "light"      # Actualiza clave existente
config["auto_save"] = True     # Agrega nueva clave
print("Actualizado:", config)

# Para eliminar
del config["lang"]
config.pop("version")          # .pop() devuelve el valor eliminado
print(" Después de eliminar:", config)

# Métodos útiles para extraer datos
claves = list(config.keys())
valores = list(config.values())
pares = list(config.items())
print("\n Claves:", claves)
print(" Valores:", valores)
print(" Pares:", pares)
```

## Relación con Tuplas: Claves Compuestas
Como vimos en la clase anterior, las tuplas son inmutables y hasheables. Esto permite usarlas como **claves compuestas** en diccionarios, algo muy útil para representar:
- Coordenadas `(x, y)` → valor
- Fechas `(año, mes, día)` → evento
- Identificadores múltiples `(id_usuario, id_recurso)` → permiso

Solo funcionan si **todos los elementos de la tupla son inmutables**. Si la tupla contiene una lista, falla el hash.

```python
# TUPLAS COMO CLAVES

# Mapa de temperaturas por ciudad y año
clima = {
    ("Santiago", 2023): 18.5,
    ("Santiago", 2024): 19.2,
    ("Valparaíso", 2023): 16.8,
    ("Valparaíso", 2024): 17.1
}

print("Temperatura en Santiago 2024:", clima[("Santiago", 2024)])

# Iteración con desempaquetado (patrón muy común en Python)
for (ciudad, año), temp in clima.items():
    print(f"  {ciudad} ({año}): {temp}°C")

# Intentamos usar una lista como clave (falla por ser mutable)
try:
    clima_invalido = {["a", "b"]: "valor"}
except TypeError as e:
    print(f"\nError esperado: {e}")
```

```python
# Claves en diccionarios (las listas NO pueden ser claves)
ubicaciones = {
    (40.7128, -74.0060): "Nueva York",
    (48.8566, 2.3522): "París"
}
print("Ubicación:", ubicaciones[(48.8566, 2.3522)])

# Iteración con índices (enumerate devuelve tuplas)
for i, valor in enumerate(("a", "b", "c")):
    print(f"Índice {i} → Valor '{valor}'")
```

###  Tuplas que contienen objetos mutables
No todas las tuplas son hasheables. Si una tupla contiene al menos un objeto mutable (como una lista, diccionario o set), **pierde su hashabilidad** porque el contenido interno podría variar.

### ¿Cómo solucionarlo?
Si necesitas empaquetar datos que originalmente son mutables, conviértelos a sus versiones inmutables antes de meterlos en la tupla:
- `list` → `tuple`
- `set` → `frozenset`

```python
# Tuplas con elementos mutables

# Tupla normal
t_valida = (1, "hola", 3.14)
print(f" t_valida es hasheable. Hash: {hash(t_valida)}")

# Tupla con lista interna
t_invalida = (1, [2, 3], 4)
try:
    print(hash(t_invalida))
except TypeError as e:
    print(f"\n  Error: {e}")

# Podemos convertir la lista interna a tupla
t_arreglada_1 = (1, tuple([2, 3]), 4)
print(f"\n  Arreglada con tuple(). Hash: {hash(t_arreglada_1)}")

#  Podemos usar frozenset para conjuntos inmutables
t_arreglada_2 = (1, frozenset([2, 3]), 4)
print(f" Arreglada con frozenset(). Hash: {hash(t_arreglada_2)}")


diccionario_final = {t_arreglada_1: "datos seguros"}
set_final = {t_arreglada_1, t_arreglada_2}
```
