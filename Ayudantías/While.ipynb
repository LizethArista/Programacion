## ¿Qué es el Bucle while?

El bucle while (mientras) es una estructura de control que **repite un bloque de código** mientras una condición sea **verdadera (True)**

### Diferencia entre if y while

| Característica | if | while |
| :--- | :--- | :--- |
| **Ejecución** | Una sola vez | Múltiples veces |
| **Condición** | Se evalúa una vez | Se evalúa en cada iteración |
| **Uso** | Tomar decisiones | Repetir acciones |


Si la condición nunca se vuelve False, el bucle nunca terminará, tendremos un **bucle infinito**.

```python
# EJEMPLO DE BUCLE INFINITO
while True:
    print("Esto nunca se detiene...")

Todo bucle while bien formado debe tener **3 elementos**:

1-. INIT (Inicialización) → Antes del while

2-. CONDITION (Condición) → En el while

3-. UPDATE (Actualización) → Dentro del while

````python
# 1. INIT - Preparamos la variable
contador = 0

# 2. CONDITION - Definimos cuándo parar
while contador <= 7:
    print(contador)

    # 3. UPDATE - Modificamos la variable para acercarnos al fin
    contador += 1
````

````python
# INIT: Inicializamos la variable de control
# Esta variable nos ayudará a saber cuántas veces se ha repetido el bucle
contador = 1

# Definimos la condición del bucle
# CONDITION: El bucle se ejecutará mientras contador sea menor o igual a 5
while contador <= 5:
    # Acción: Código que se repite en cada iteración
    print(f"Iteración número: {contador}")

    # UPDATE: actualizar la variable de control (si no se actualiza, será infinito)
    contador = contador + 1  # También se puede escribir como: contador += 1

# Código que se ejecuta DESPUÉS de que el bucle termina
print(f"Valor final del contador: {contador}")
````

````python
# while con else

# El bloque 'else' se ejecuta cuando el while termina NORMALMENTE
# (no cuando se usa 'break')

contador = 1

while contador <= 3:
    print(f"Iteración {contador}")
    contador += 1
else:
    # Este código se ejecuta UNA VEZ cuando el while termina
    print("Conteo finalizado")
````

## break y continue

###   `break` - Romper el Bucle
Detiene el bucle **inmediatamente**, incluso si la condición sigue siendo verdadera

###   `continue` - Saltar Iteración
Salta el resto del código en la iteración actual y **vuelve al inicio** del bucle

###   Diferencia Visual

while condicion:

código...

if algo:

break → SALE del bucle completamente

continue → VUELVE al inicio del while (salta lo que sigue)

código... → (este no se ejecuta si hay continue)


````python
# Uso de break - Búsqueda con límite

# Buscamos un número específico, pero máximo 10 intentos
numero_secreto = 7
intento = 0
max_intentos = 10

print(f"  Adivina el número (1-10). Tienes {max_intentos} intentos")

while intento < max_intentos:
    # Solicitamos entrada al usuario
    adivinanza = int(input("Tu número: "))

    # Aumentamos el contador de intentos
    intento += 1

    # Verificamos si acertó
    if adivinanza == numero_secreto:
        print(f"  Adivinaste el número en {intento} intento(s)")
        break  # SALIMOS del bucle inmediatamente
    else:
        print(f" Incorrecto. Te quedan {max_intentos - intento} intentos")

# Este código se ejecuta después del break o cuando termina el while
if intento == max_intentos and adivinanza != numero_secreto:
    print(f"  Game Over. El número era {numero_secreto}")
````

````python
# Uso de continue - Saltar valores

# Imprimimos solo los números IMPARES del 1 al 10
# Usamos continue para saltar los números pares

numero = 0

print("Números impares del 1 al 10:")

while numero < 10:
    numero += 1  # Aumentamos

    # Si el número es par, saltamos esta iteración
    if numero % 2 == 0:
        continue  # Salta al inicio del while, no ejecuta el print

    # Este código solo se ejecuta si el número es impar
    print(f"{numero}")
print("Fin de la lista")

````


### Otra función útil... is.digit()

`.isdigit()` es un **método de strings** que verifica si **todos los caracteres** en una cadena son **dígitos numéricos**.

###   Sintaxis
```python
texto.isdigit()  # Devuelve True o False

útil para validar entradas de usuario antes de convertir a número con int()

````python
#   SIN validación
edad = input("Edad: ")
edad_numero = int(edad)
````

````python
#   CON validación
edad = input("Edad: ")
if edad.isdigit():
    edad_numero = int(edad)
else:
    print("Entrada no válida")
````

Podemos usar `while` para **seguir pidiendo** datos hasta que el usuario ingrese algo válido


````python
while True:
    edad = input("Ingresa tu edad: ")

    if edad.isdigit():  # Validación
        break  # Salimos si es válido
    else:
        print("Intenta de nuevo")
````
