**1. Operadores Aritméticos Básicos**

En Python, podemos usar variables como una calculadora. Estos son los operadores más comunes:

$+$ : Suma

$-$ : Resta

$ *$ : Multiplicación

$/$ : División (siempre devuelve un número decimal float)

$//$ : División entera (devuelve solo la parte entera)

$\%$ : Módulo (devuelve el resto de la división)

$**$ : Potencia (ej: 2**3 es 2 elevado a 3)

```python
# Vamos a definir dos variables y con ellas realizaremos operaciones

a = 10
b = 3

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División normal:", a / b)
print("División entera:", a // b)
print("Módulo (Resto):", a % b)
print("Potencia:", a ** b)

#  Cambia los valores de 'a' y 'b' arriba y vuelve a ejecutar.
# ¿Qué pasa si 'b' es 0 en la división?


```

**2. Precedencia de Operadores (El orden importa)**

Al igual que en matemáticas, Python tiene un orden para evaluar las expresiones:

Paréntesis (): Tienen la máxima prioridad.

Potencia **.

Multiplicación *, División /, Módulo %.

Suma + y Resta -.

Si hay operadores del mismo nivel, se evalúan de izquierda a derecha.


```python
# ¿Cuál es el resultado?
resultado_1 = 2 + 3 * 4
print("2 + 3 * 4 =", resultado_1)

resultado_2 = (2 + 3) * 4
print("(2 + 3) * 4 =", resultado_2)



resultado_3 = 10 / 2 + 3
print("10 / 2 + 3 =", resultado_3)

resultado_4 = 10 / (2 + 3)
print("10 / (2 + 3) =", resultado_4)
```

**3. Interactuando con el Usuario (input)**

Hasta ahora, los valores los escribimos nosotros en el código. Pero, ¿qué pasa si queremos que el usuario escriba los datos?

Usamos la función *input("Mensaje")*

*IMPORTANTE:* input() siempre devuelve un TEXTO (string).

Si queremos hacer matemáticas con lo que escribe el usuario, debemos convertirlo a número usando int() (entero) o float() (decimal).

```python
edad = input("¿Cuántos años tienes?") # Esto es texto "20"
print(type(edad))

edad_numero = int(edad) # Esto es número 20
print("Tu edad es: ", edad_numero)
```

```python
# Veamos más ejemplos....
nombre = input("¿Cuál es tu nombre? ")
print("¡Hola,", nombre, "!")

# Ahora probemos con números... es importante el tipo de dato
numero1 = int(input("Ingresa un número entero: "))
numero2 = int(input("Ingresa otro número entero: "))

suma = numero1 + numero2
print("La suma de tus números es:", suma)
```

**Errores Comunes**

1. TypeError: can only concatenate str (not "int") to str:

Causa: Estás intentando sumar un texto con un número (ej: "Tu edad es: " + 20).

Solución: Convierte el número a texto str(20) o usa comas en el print.


2. ValueError: invalid literal for int() with base 10:

Causa: Intentaste convertir a int() un texto que no es número (ej: int("hola") o int("10.5")).

Solución: Usa float() si esperas decimales, o asegúrate de que el usuario escriba solo números.

3. El programa se queda "corriendo":
Causa: Estás esperando que el usuario escriba en el input(). Busca el cuadro de texto en la celda.


```python
# Calculadora de Propinas
# Vamos a combinar todo: Input, conversión de tipos y operadores

print("--- CALCULADORA DE CUENTA ---")

# 1. Pedir el monto de la cuenta
monto_cuenta = float(input("Ingresa el monto total de la cuenta ($): "))

# 2. Pedir el porcentaje de propina que quieren dar
porcentaje_propina = float(input("Ingresa el porcentaje de propina (ej: 10, 15, 20): "))

# 3. Calcular el valor de la propina
# Fórmula: (monto * porcentaje) / 100
valor_propina = (monto_cuenta * porcentaje_propina) / 100

# 4. Calcular el total a pagar
total_pagar = monto_cuenta + valor_propina

# 5. Mostrar resultados
print("\n--- RESUMEN ---")
print("Monto original: $", monto_cuenta)
print("Propina (" + str(porcentaje_propina) + "%): $", valor_propina)
print("TOTAL A PAGAR: $", total_pagar)

# Ejercicio:
# Modifica el código para que también pida cuántas personas son
# y calcule cuánto debe pagar cada uno
```

**4. Condicionales: if, elif, else**

¿Qué pasa si queremos que el programa tome decisiones?

Imagina que eres un guardia de seguridad en un club:

SI (if) tienes 18 años o más ➡️ Entras

SI NO (else) ➡️ No entras

En Python, usamos las palabras clave if, elif (else if) y else para esto

**NOTAS IMPORTANTES:**

1. Dos puntos-. : Al final de la línea del if o else, siempre debes poner dos puntos.


2. Indentación (Sangría)-.  El código que está dentro del if debe estar indentado (generalmente 4 espacios o un tab). Python usa esto para saber qué líneas pertenecen a la condición

3. Comparación: Para comparar si dos valores son iguales, usamos == (doble igual).

Un solo = es para asignar valores

```python
# Ejemplo 1: Verificando una edad fija
edad = int(input("Ingresa tu edad: "))

print("Verificando acceso...")

if edad >= 18:
    print(" Eres mayor de edad. ¡Puedes pasar!")
    print(" Disfruta la fiesta")
else:
    print(" Eres menor de edad, no puedes pasar")
    print(" Vuelve a casa")

print("--- Fin del programa ---")
```

¿Qué acabamos de ver?

Python evalúa la condición edad >= 18.
Si es Verdadero (True): Ejecuta las líneas indentadas bajo el if

Si es Falso (False): Salta esas líneas y ejecuta las que están bajo el else

```python
# Ejemplo 3: Semáforo
# El usuario elige un color y el programa dice qué hacer

color = input("¿De qué color está el semáforo? (rojo/amarillo/verde): ").lower()

if color == "rojo":
    print("🛑 ¡ALTO! No cruces")
elif color == "amarillo":
    print("⚠️ Precaución, prepara para parar o pasar con cuidado")
elif color == "verde":
    print("🟢 ¡Avanza! Puedes cruzar")
else:
    print("Color no reconocido. Por favor escribe rojo, amarillo o verde")
```
